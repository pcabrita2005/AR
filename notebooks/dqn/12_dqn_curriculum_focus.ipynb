# Notebook 12: DQN Curriculum Focus

Este notebook aprofunda uma agenda concreta de curriculum DQN sem misturar a comparacao global do notebook `11`.


## Objetivo

O objetivo deste notebook e:

- analisar uma agenda especifica por seed;
- observar as curvas `vs_random`, `vs_weak`, `vs_strong` e `vs_previous`;
- comparar checkpoints e renderizar uma partida de exemplo.


```python
from __future__ import annotations

import copy
import json
import sys
from pathlib import Path

import matplotlib.pyplot as plt
import torch

ROOT = Path.cwd().resolve()
while not (ROOT / "config.yaml").exists() and ROOT != ROOT.parent:
    ROOT = ROOT.parent
if not (ROOT / "config.yaml").exists():
    raise FileNotFoundError("Could not find project root containing config.yaml")
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

OUTPUTS = ROOT / "outputs"
CURRICULUM_OUTPUTS = OUTPUTS / "dqn_curriculum_experiments"
CURRICULUM_OUTPUTS.mkdir(parents=True, exist_ok=True)

from connect4_rl.agents.baselines import StrongHeuristicAgent, WeakHeuristicAgent
from connect4_rl.envs.connect_four import apply_action, initial_state, is_terminal, legal_actions, render_ascii
from connect4_rl.experiments import build_agent_from_checkpoint, compute_elo_ratings, round_robin_detailed
from connect4_rl.experiments.dqn_curriculum import build_default_dqn_curricula, expand_curriculum_schedule, train_dqn_with_curriculum
from connect4_rl.config import load_config

CONFIG = load_config(ROOT / "config.yaml")
NOTEBOOK_DEVICE = CONFIG.resolve_device()
if NOTEBOOK_DEVICE == "cuda":
    torch.set_default_device(NOTEBOOK_DEVICE)
print({"device": NOTEBOOK_DEVICE, "outputs": str(CURRICULUM_OUTPUTS)})
```

## Passo 1: Definicao da agenda em foco


```python
run_training = False

curricula = build_default_dqn_curricula()
focus_agenda = "curriculum_classic"
seeds = [7, 17, 27]

focus_config = copy.deepcopy(CONFIG)
focus_config.global_.device = NOTEBOOK_DEVICE
focus_config.dqn.episodes = 180
focus_config.dqn.eval_interval = 30
focus_config.dqn.eval_games = 12
focus_config.dqn.gradient_updates_per_step = 2

{"focus_agenda": focus_agenda, "seeds": seeds, "episodes": focus_config.dqn.episodes}
```

## Passo 2: Vista da agenda


```python
definition = curricula[focus_agenda]
schedule, phase_summary = expand_curriculum_schedule(focus_config.dqn.episodes, definition, seed=seeds[0])
print(definition.description)
for item in phase_summary:
    print(item)
```

## Passo 3: Treino opcional da agenda em foco


```python
if run_training:
    for seed in seeds:
        run_config = copy.deepcopy(focus_config)
        run_config.global_.seed = seed
        checkpoint_dir = CURRICULUM_OUTPUTS / f"{focus_agenda}_seed_{seed}"
        print(f"=== Running {focus_agenda} | seed={seed} ===")
        _agent, metrics = train_dqn_with_curriculum(definition, run_config, checkpoint_dir=checkpoint_dir)
        print("Last eval:", metrics.evaluation[-1] if metrics.evaluation else {})
else:
    print("Training skipped. Set run_training = True to launch this agenda.")
```

## Passo 4: Carregamento das corridas da agenda


```python
def load_focus_runs(root: Path, family: str) -> list[dict]:
    runs = []
    for metrics_path in sorted(root.glob(f"{family}_seed_*/metrics_final.json")):
        seed = int(metrics_path.parent.name.rsplit("_", 1)[-1])
        data = json.loads(metrics_path.read_text(encoding="utf-8"))
        runs.append({"seed": seed, "metrics_path": metrics_path, "data": data})
    return runs

focus_runs = load_focus_runs(CURRICULUM_OUTPUTS, focus_agenda)
[(run["seed"], len(run["data"].get("evaluation", []))) for run in focus_runs]
```

## Passo 5: Curvas por seed


```python
if focus_runs:
    fig, axes = plt.subplots(2, 2, figsize=(14, 8), sharex=True)
    axes = axes.ravel()
    for run in focus_runs:
        seed = run["seed"]
        evaluations = run["data"].get("evaluation", [])
        if not evaluations:
            continue
        episodes = [item["episode"] for item in evaluations]
        axes[0].plot(episodes, [item["vs_random_win_rate"] for item in evaluations], marker="o", label=f"seed {seed}")
        axes[1].plot(episodes, [item["vs_weak_heuristic_win_rate"] for item in evaluations], marker="o", label=f"seed {seed}")
        axes[2].plot(episodes, [item["vs_strong_heuristic_win_rate"] for item in evaluations], marker="o", label=f"seed {seed}")
        axes[3].plot(episodes, [item["vs_previous_win_rate"] for item in evaluations], marker="o", label=f"seed {seed}")

    axes[0].set_title("vs random")
    axes[1].set_title("vs weak")
    axes[2].set_title("vs strong")
    axes[3].set_title("vs previous snapshot")
    for ax in axes:
        ax.legend()
        ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
else:
    print("No runs found for this agenda.")
```

## Passo 6: Torneio entre seeds


```python
def build_seed_agents(runs: list[dict]) -> dict[str, object]:
    agents = {}
    for run in runs:
        checkpoint = run["data"].get("best_checkpoint_path")
        config = run["data"].get("config", {})
        if not checkpoint:
            continue
        agents[f"{focus_agenda}_seed_{run['seed']}"] = build_agent_from_checkpoint(
            "dqn",
            ROOT / checkpoint,
            config,
            device=NOTEBOOK_DEVICE,
        )
    return agents

seed_agents = build_seed_agents(focus_runs)
list(seed_agents)
```

```python
if len(seed_agents) >= 2:
    match_log = round_robin_detailed(seed_agents, games_per_matchup=8)
    ratings = compute_elo_ratings(match_log)
    print("Elo:", ratings)
    match_log[:3]
else:
    print("Need at least two trained seeds with checkpoints.")
```

## Passo 7: Visualizacao de uma partida


```python
def play_and_render(agent, opponent, controlled_player: int = 1) -> str:
    state = initial_state()
    frames = [render_ascii(state)]
    move_idx = 0
    while not is_terminal(state):
        if state.current_player == controlled_player:
            action = agent.select_action(state, legal_actions(state))
            actor = agent.name
        else:
            action = opponent.select_action(state, legal_actions(state))
            actor = opponent.name
        state = apply_action(state, action)
        move_idx += 1
        frames.append(f"Move {move_idx}: {actor} played column {action}")
        frames.append(render_ascii(state))
    frames.append(f"Winner: {state.winner}")
    return "".join(frames)

if seed_agents:
    sample_name = sorted(seed_agents)[0]
    sample_agent = seed_agents[sample_name]
    print(sample_name)
    print(play_and_render(sample_agent, StrongHeuristicAgent(seed=123), controlled_player=1))
else:
    print("No trained agent available yet.")
```

## Leituras esperadas

- se a mesma agenda variar demasiado entre seeds, o curriculum pode estar sensivel demais ao replay inicial;
- se o agente renderizado ainda falhar bloqueios obvios contra `strong`, o shaping nao esta a produzir transferencia suficiente;
- se o torneio interno destacar claramente uma seed, vale a pena usar esse checkpoint no notebook de comparacao global.
